find the highest percentage scored in each category and display the categories in descending order of percentage."

```python
z=df.groupby('Category',as_index=False)['Percentage'].max().sort_values('Percentage',ascending=False)
z
```

```python
sns.set(rc={'figure.figsize':(12,6)})
x=sns.barplot(x='Category',y='Percentage',hue='Category',data=z,legend=False)
for bars in x.containers:
    x.bar_label(bars)
```

calculate the total marks obtained by each gender and display the results in descending order of total marks."

```python
z=df.groupby('Gender',as_index=False)['Total Marks'].sum().sort_values(by='Total Marks',ascending=False)
z
```

```python
sns.set(rc={'figure.figsize':(12,6)})
x=sns.barplot(x='Gender',y='Total Marks',hue='Gender',data=z)
for bars in x.containers:
    x.bar_label(bars)
```

calculate the total marks obtained by students in each standard from Faridabad city.

```python
z=df[(df['City']=='Faridabad')].groupby('Standard')['Total Marks'].sum().reset_index(name='Count')
```

```python
z
```

```python
plt.figure(figsize=(10,6))
z.plot(x='Standard',y='Count',kind='bar',color=['red','pink','black'])
plt.xlabel('Standard')
plt.ylabel('count')
plt.show()
```

count the number of students in each category who scored more than 70%.

```python
z=df[df['Percentage']>70].groupby('Category')['Percentage'].count().reset_index(name='count')
```

```python
z
```

```python
sns.set(rc={'figure.figsize':(12,5)})
x=sns.barplot(x='Category',y='count',hue='Category',data=z)
for bars in x.containers:
    x.bar_label(bars)
```

filters the dataset to include only those students who scored more than 70% in the Percentage column.

```python
z=df[df['Percentage']>70].groupby('Category')['Percentage'].count()
plt.figure(figsize=(12,6))
z.plot(kind='pie',shadow=True,autopct='%1.1f%%',startangle=120,explode=[0.1,0,0.1,0.1])
plt.xlabel('Standard')
plt.ylabel('count')
plt.show()
```

count the number of students in each category for every admission year."

```python
z=df.groupby(['Admission Year','Category']).size().reset_index(name='count')
```

```python
z
```

```python
sns.set(rc={'figure.figsize':(12,6)})
x=sns.barplot(x='Admission Year',y='count',hue='Category',data=z)
for bars in x.containers:
    x.bar_label(bars)
```
